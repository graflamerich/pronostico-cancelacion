## introducción

Interconnet, el operador de telecomunicaciones, proporciona diferentes servicios a sus clientes. Principalmente planes de internet y comunicación por teléfono fijo a través de la conexión a varias líneas. También, los usuarios se benefician con otros servicios como seguridad en línea, almacenamiento de datos en la nube, backup de datos, línea de soporte técnico, streaming de TV y películas. Interconect ofrece varias metologías de pago: mensual o un contrato anual o bianual, utilizando varios métodos de pago y recibir una factura electrónica después de una transacción.

La intención de este estudio es predecir si el cliente tiene intención de suspender los servicios contratados, para de esa manera poderle ofrecer un plan que se ajuste a sus necesidades y evitar el éxodo de clientes. Apoyados en el equipo de marketing quienes proporcionaron las 4 bases de datos usadas en este estudio, esperan recibir nuestras respuestas, luego de construir el mejor modelo de clasificación, para poder ofrecerle a estos clientes una mejor opción. 

La estructura del estudio consta de varias secciones: comenzaremos por la importación de las librerías a utilizar y los 4 datasets sumistrados. De cada base de datos, haremos un desglose de datos, obteniendo valores ausentes, duplicados, número de entradas, datos en común en las tablas par poder unificarlas y hacer el estudio exhaustivo que la empresa espera. Estarán incluidos los gráficos correspondientes, los análisis y metodologías de evaluación del modelo. Luego los modelos y finalmente los resultados y un informe completo.

Crearemos una función para convertir nombres de columnas de DataFrame de camelCase a snake_case la cual aplicaremos a todos los data frames.

```python
_underscorer1 = re.compile(r'(.)([A-Z][a-z]+)')
_underscorer2 = re.compile('([a-z0-9])([A-Z])')

def to_snake_case(columns):
    
    return list(map(lambda string: _underscorer2.sub(r'\1_\2', _underscorer1.sub(r'\1_\2', string)).lower(), columns))
```

## evaluación

En esta sección del proyecto determinaremos cual será la función que utilizaremos para evaluar todos los modelos que estudiaremos. Tomando en cuenta los parámetros estipulados: Métrica principal: AUC-ROC y métrica adicional: exactitud.

**`model_gridsearch`**: devolverá los mejores hiperparámetros y métrica AUC-ROC obtenidos a partir de un ajuste con GridSearchCV.

**`evaluate_model`**: devolverá las métricas AUC-ROC y exactitud para el conjunto de entrenamiento y prueba, así como también  la curva AUC-ROC, curva sensibilidad-precisión y una matriz de confusión.

**`cf_matrix`**: devolverá una matriz de confusión para el conjunto de entrenamiento y prueba.

**`feature_vip`**: devolverá gráficos de barras para visualizar la importancia de variables en la construcción de los diferentes modelos.

```python
def model_gridsearch(estimator_name, params, x_train, y_train, model_name):
    
    '''
    Devuelve los mejores hiperparámetros y el RECM más bajo para un determinado algoritmo de aprendizaje.
    
    : estimator_name: nombre del algoritmo
    : params: hiperparámetros de cada modelo
    : x_train: características entrenamiento
    : y_train: objetivo entrenamiento.
    
    '''
    grid_search= GridSearchCV(
    estimator=estimator_name,
    param_grid=params,
    scoring= 'roc_auc',
    cv=3)

    grid_search.fit(x_train, y_train)

    max_score = grid_search.cv_results_['mean_test_score'].max()
    index_max_score = np.where(grid_search.cv_results_['mean_test_score'] == max_score)[0][0]

    best_set_of_params = grid_search.cv_results_['params'][index_max_score]
    
    print(f'Mejores Hiperparámetros {model_name}: {best_set_of_params}')
    print(f'Mejor AUC-ROC {model_name}: {max_score}')
    
    return 
```

```python
def evaluate_model(model, train_features, train_target, test_features, test_target, model_name):
    
    '''
    Devuelve los valores de exactitud, ROC-AUC. 
    También devuelve gráficos de curva ROC-AUC, F1 y PRC (precision-recall curve)
    
    :model: modelo previamente entrenado y con ajuste de hiperparámetros
    :train_features: características de entrenamiento
    :train_target_ características de prueba
    :test_features: objetivo de entrenamiento
    :test_target: objetivo de prueba
    
    '''
    eval_stats = {}
    
    fig, axs = plt.subplots(1, 3, figsize=(20, 6)) 
    
    for type, features, target in (('train', train_features, train_target), ('test', test_features, test_target)):
        
        eval_stats[type] = {}
    
        pred_target = model.predict(features)
        pred_proba = model.predict_proba(features)[:, 1]
        
        # F1
        f1_thresholds = np.arange(0, 1.01, 0.1)
        f1_scores = [metrics.f1_score(target, pred_proba>=threshold) for threshold in f1_thresholds]
        
        # ROC
        fpr, tpr, roc_thresholds = metrics.roc_curve(target, pred_proba)
        roc_auc = metrics.roc_auc_score(target, pred_proba)    
        eval_stats[type]['ROC AUC'] = roc_auc

        # PRC
        precision, recall, pr_thresholds = metrics.precision_recall_curve(target, pred_proba)
        aps = metrics.average_precision_score(target, pred_proba)
        eval_stats[type]['APS'] = aps
        
        if type == 'train':
            color = 'blue'
        else:
            color = 'green'

        # Valor F1
        ax = axs[0]
        max_f1_score_idx = np.argmax(f1_scores)
        ax.plot(f1_thresholds, f1_scores, color=color, label=f'{type}, max={f1_scores[max_f1_score_idx]:.2f} @ {f1_thresholds[max_f1_score_idx]:.2f}')
        # establecer cruces para algunos umbrales        
        for threshold in (0.2, 0.4, 0.5, 0.6, 0.8):
            closest_value_idx = np.argmin(np.abs(f1_thresholds-threshold))
            marker_color = 'orange' if threshold != 0.5 else 'red'
            ax.plot(f1_thresholds[closest_value_idx], f1_scores[closest_value_idx], color=marker_color, marker='X', markersize=7)
        ax.set_xlim([-0.02, 1.02])    
        ax.set_ylim([-0.02, 1.02])
        ax.set_xlabel('threshold')
        ax.set_ylabel('F1')
        ax.legend(loc='lower center')
        ax.set_title(f'Valor F1') 

        # ROC
        ax = axs[1]  
        ax.plot(fpr, tpr, color=color, label=f'{type}, ROC AUC={roc_auc:.2f}')
        # establecer cruces para algunos umbrales        
        for threshold in ( 0.2, 0.4, 0.5, 0.6, 0.8):
            closest_value_idx = np.argmin(np.abs(roc_thresholds-threshold))
            marker_color = 'orange' if threshold != 0.5 else 'red'            
            ax.plot(fpr[closest_value_idx], tpr[closest_value_idx], color=marker_color, marker='X', markersize=7)
        ax.plot([0, 1], [0, 1], color='grey', linestyle='--')
        ax.set_xlim([-0.02, 1.02])    
        ax.set_ylim([-0.02, 1.02])
        ax.set_xlabel('FPR')
        ax.set_ylabel('TPR')
        ax.legend(loc='lower center')        
        ax.set_title(f'Curva ROC')
        
        # PRC
        ax = axs[2]
        ax.plot(recall, precision, color=color, label=f'{type}, AP={aps:.2f}')
        # establecer cruces para algunos umbrales        
        for threshold in (0.2, 0.4, 0.5, 0.6, 0.8):
            closest_value_idx = np.argmin(np.abs(pr_thresholds-threshold))
            marker_color = 'orange' if threshold != 0.5 else 'red'
            ax.plot(recall[closest_value_idx], precision[closest_value_idx], color=marker_color, marker='X', markersize=7)
        ax.set_xlim([-0.02, 1.02])    
        ax.set_ylim([-0.02, 1.02])
        ax.set_xlabel('recall')
        ax.set_ylabel('precision')
        ax.legend(loc='lower center')
        ax.set_title(f'PRC')  
        
        eval_stats[type]['Exactitud'] = metrics.accuracy_score(target, pred_target)
        eval_stats[type]['F1'] = metrics.f1_score(target, pred_target)
        eval_stats[type]['Precisión'] = metrics.precision_score(target, pred_target)
        eval_stats[type]['Sensibilidad'] = metrics.recall_score(target, pred_target)

    df_eval_stats = pd.DataFrame(eval_stats)
    df_eval_stats = df_eval_stats.round(4)
    df_eval_stats = df_eval_stats.reindex(index=('Exactitud', 'Sensibilidad', 'Precisión', 'F1', 'APS', 'ROC AUC'))
    
    print(f'{model_name}:')
    print(df_eval_stats)

    return
```

```python
def cf_matrix(model, train_features, train_target, test_features, test_target, model_name):
    
    '''
    Devuelve una matriz de confusión para el conjunto de entrenamiento y prueba
    
    :model: modelo previamente entrenado y con ajuste de hiperparámetros
    :train_features: características de entrenamiento
    :train_target_ características de prueba
    :test_features: objetivo de entrenamiento
    :test_target: objetivo de prueba
    
    '''
    # Definimos el número y tamaño de subplots
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))
    
    # Construimos una matriz de confusión para el conjunto de entrenamiento
    ax = axs[0]
    plot_confusion_matrix(model, train_features, train_target,
                                 display_labels=[0,1],
                                 cmap=plt.cm.YlOrRd,
                                 ax=ax)
    ax.set_title('Entrenamiento')
    ax.grid(False)
    
    # Construimos una matriz de confusión para el conjunto de prueba
    ax = axs[1]
    plot_confusion_matrix(model, test_features, test_target,
                                 display_labels=[0,1],
                                 cmap=plt.cm.YlOrRd,
                                 ax=ax)
    ax.set_title('Prueba')
    ax.grid(False)
    
    plt.suptitle(f'Matrices de confusión {model_name}')
    
    return 
```

```python
def feature_importance(model, x_train, model_name):
    
    '''
    Devuelve gráficos de barras de la importancia de características por modelo

    :model: nombre de modelo previamente entrenado y con ajuste de hiperparámetros
    :x_train: características de entrenamiento
    :model_name: nombre del modelo para título de gráfico

    '''

    if model == lr_model:
    
        # Obtenemos los coeficientes de importancia de modelo de regresión logística
        lr_model_importance = pd.DataFrame(
            model.coef_[0], index=x_train.columns, columns=['Importancia']
        )
        # Trazamos un gráfico de barras
        lr_model_importance.sort_values(by='Importancia').plot(kind='bar', color='darkred', figsize=(15, 6))
        plt.xticks(rotation=45, ha='right', rotation_mode='anchor', fontsize=13)
        plt.title('Importancia de características Modelo de Regresión Logística')
        plt.show()
        
    else:
        
        # Obtneemos valores de importancia de otros modelos a partir de feature_importances
        sorted_feature_importance = model.feature_importances_.argsort()
        plt.figure(figsize=(8,6))
        plt.barh(x_train.columns[sorted_feature_importance], 
             model.feature_importances_[sorted_feature_importance], 
             color='darkred')
        plt.title(f'Importancia de características {model_name}')
        
    return
```
